Through normalization we can convert these features in normalised form

```python
## We can use power transformer to standardise the data 
from sklearn.preprocessing import PowerTransformer
```

```python
# Initialize PowerTransformer
power_transformer = PowerTransformer(method='box-cox', standardize=True)
```

```python
# Fit and transform the skewed data
transformed_data = power_transformer.fit_transform(df[['total sulfur dioxide']])
```

```python
df['total sulfur dioxide'] = pd.DataFrame(transformed_data)
```

```python
## PCA 

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame with the features you provided
# Replace 'target_column' with the column containing the target variable if applicable
target_column = 'quality'
X = df.drop(columns=[target_column])  # Exclude the target variable from the features

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Determine the number of components to keep
explained_variance_ratio_cumsum = pca.explained_variance_ratio_.cumsum()
num_components_to_keep = (explained_variance_ratio_cumsum <= 0.95).sum()

# Apply PCA with the chosen number of components
pca_final = PCA(n_components=num_components_to_keep)
X_pca_final = pca_final.fit_transform(X_standardized)

# Create a DataFrame with the principal components
columns_pca = [f'PC{i+1}' for i in range(num_components_to_keep)]
df_pca = pd.DataFrame(data=X_pca_final, columns=columns_pca)

# Concatenate the principal components DataFrame with the target variable
df_result = pd.concat([df[target_column], df_pca], axis=1)

# Display the result
print(df_result.head())
```
